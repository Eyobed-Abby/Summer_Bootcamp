#### ***`Let's learn how to print patterns like`***
```python
1
12
123
1234
12345
```

`As you might have guessed, we use nested loop for such patterns`

Any idea how to set things up?

let's break the problem down into two sections
1. We need five rows(top to bottom)
2. In each row, we need to print numbers depending on the row number

Meaning: first row has 1, second row has 12.... fifth row has 12345

let's address the first section(printing five rows with just 1)

```python
for i in range(1, 6): #remember range function this means start at 1 and stop at 5(6-1)
  print(i)
```

````python
for i in range(1, 6):
  print(i)
````

Great now we have five rows

What's missing?

we want for each row to print numbers from 1 to that row number!
* this means for 1st row we just want 1
* for second row we go from 1 - 2
* for third row we go from 1 - 3
* for fourth row we go from 1 - 4
* for the fifth row we go from 1- 5

Any idea on what's going on?

yes! we are looping again in every interation

**In each row we iterate from 1 to the row number before moving to the next row!**

For example:
if i is 2, it means we are in second row so we need to print 12, how do we do that?

we add loop going from 1 to 2 and print

```python
for i in range(1, 6):
  #add another loop to go from 1 to i
  for j in range(1, i): #what is wrong here? we miss i right?
    print(j)
```

`Let's try it out`

````python
for i in range(1, 6):
  #add another loop to go from 1 to i
  for j in range(1, i): #what is wrong here? we miss i right?
    print(j)
````

**What's wrong with the above code?**
1. We are printing eveyrthing down... we need to make sure that same row numbers should be together
2. we are missing fifth row, why? this is because or our range function range(1,i) mean it doesn't include i so our range function should be from (1, i+1)



let's fix the firt problem and learn one more thing about print() function.

* print() function has an argument called `"end"` and by default the value is new line, which means as soon as print is executed it goes to the next line
* that's why in the above code instead of 123 we get 1 2 and 3 in different lines
* to fix that problem we change the value of end in the print() function to empty string `""` like `end = ""`


The updated code would look like:
```python
for i in range(1, 6):
  for j in range(1, i):
    print(j, end="")
```

````python
for i in range(1, 6):
  for j in range(1, i):
    print(j, end="")
````

* Now we fixed the issue but we got another one, everything is getting printed in one line
so we need to make sure it goes to the next line for each row. so we will just add normal print line for each outer loop like:
```python
for i in range(1, 6):
  for j in range(1, i):
    print(j, end="")
  
  print()#this print will just move the cursor to new line
```

````python
for i in range(1, 6):
  for j in range(1, i):
    print(j, end="")

  print()#this print will just move the cursor to new line
````

**Perfect!**

Now let's fix the second issue, which is about the range function.

remember the range function doesn't inlcude the end value it stops at end-1

so if the range is from range(1,5) means from 1 to 4, so to fix that we need to add one to the end value.

```python
for i in range(1, 6):
  for j in range(1, i+1): #we add one to the range function end
    print(j, end="")
  
  print()
```

````python
for i in range(1, 6):
  for j in range(1, i+1):
    print(j, end="")

  print()
````

#### ***`Let's learn how to build the following pattern together`***
```python
   *
  ***
 *****
*******
```

What's differnt from the previous pattern?
1. There is only odd number of starts every row 1, 3, 5, 7...
2. There are leading spaces in every row but the last one

How can we implement this?
1. First we try to come up in printing the odd numbered stars like:

```python
*
***
*****
*******
```
2. we add space to the left. But remember the number of spaces decrease in every row. since we have 4 rows, the first row has 3 spaces, second row has 2 spaces, third row has 1 space and last one has no space.

`see any pattern?` For every row i we have n-i spaces, where n is the total number of rows

for example: in the above example we have 4 rows so n = 4

```python
first row i = 1, so 4-1 spaces = 3
second row i = 2, so 4-2 spaces = 2
third row i = 3, so 4-3 spaces = 1
fourht row i = 4, so 4-4 spaces = 0
```


````python
#let's make the traingles first
for i in range(1, 5):
  for j in range(1, i*2): #remember we always have odd number of stars that why we have i*2 as stopping position
    print("*", end="")
  print()
````

````python
#let's add the leading spaces before the stars
for i in range(1, 5):
  #here we need to add the spaces
  #how many spaces? 4-i in this case
  for k in range(4-i):
    print(" ", end="")
  for j in range(1, i*2):
    print("*", end="")
  print()
````
