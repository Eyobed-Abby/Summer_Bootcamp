# Using 'else' clause in loops

* When the loop condition of `"for" or "while" statement fails then code part in "else" is executed`

* **`If a break statement is executed inside the for loop then the "else" part is skipped.`**

* **`Note that the "else" part is executed even if there is a continue statement.`**

````python
# Prints out 0,1,2,3,4 and then it prints "count value reached 5"

count=0
while(count<5):
    print(count)
    count +=1
else:
    print(f"count value reached {count}")
````

````python
# Prints out 1,2,3,4
for i in range(1, 10):
    if(i%5==0):
        break
    print(i)
else:
    print("this is not printed because for loop is terminated because of break but not due to fail in condition")
````

# Functions

A function is a block of code which only runs when it is called.

* Convenient way to divide our code into useful blocks
* Allowing us to order our code
* Make it more readable
* Reuse it and save some time

Functions in python are defined using the block keyword **"def"**, followed with the function's name

```python
    def function_name(arguments if any):
        component of the function
```

````python
#declaring a function named "function_name"
def function_name(): #it has no arguments
  print("Hello from a function") #this is what hte function is doing, printing

function_name() #calling the function
````

````python
def call_name():
    print('Hello,Alem')
call_name()
````

### Remember function runs only when it is called

````python
# let's call our function
function_name() # we are calling our function
````

````python
call_name()
````

## Types of Functions

There are two types of functions in Python:
### 1. **Built-in functions**:
These are the functions that are already built-in in Python such as:

  * help() to ask for help,
  * min() to get the minimum value,
  * print() to print an object to the terminal,
  * range() to get numbers in a given range.

````python
# getting the absolute value
abs(-10)
````

````python
# getting input from user
input("hello, please insert your name: ")
````

````python
# finding maximum and minimum
l = [2, -9, 4]
max(l), min(l)
````

### 2. **User-Defined Functions (UDFs):**
These are the functions **`created by the user`** according to their programming needs.

Let's create a function that takes two numbers as arguments and adds them and returns the sum

````python
def sum(a, b):   # This is a user defined function that takes two arguments
    return a + b # and adds them up.
````

**`Rememeber again inorder to use a function, we need to call it,`** so let's call it

````python
sum(2, 5) #takes two numbers a =2 and b =5 inthis case and returns their sum which is 7
````

### Every function takes 0 or more arguments

* Information can be passed into functions as arguments.

* Arguments are specified after the function name, inside the parentheses.

* You can add as many arguments as you want, just separate them with a comma.

````python
def func(argument):  # as you can see, our function 'func' has argument named 'argument'
    print(argument)   # the job of this function is to print the argument
````

````python
# Let's call our function func to see what it does
func()  # if we call it like this, we get an error because func has argument in it
````

````python
func('hello') # now we are correct as we specified our argument
````

````python
func(1) # whatever we put for the arguemnt will be printed
````

#### Anything after "return" will be ignored!! afer return the function stops running

````python
def sum(a, b):
    return a + b
    print(b)   # this print statement is after the return statemnet, so it will be ignored

sum(3,9) # it will just retun the sum, will not print 9
````

````python
# to print 9, we should make it above the return statement
def sum(a, b):
    print(b)  # now this one will be executed
    return a + b

sum(3, 9)
````
