### Adding Items into List
we use `.append()` function to add element at the end of the list

````python
mylist = [1, 2, 3]
# so we want to add an element 4 in our list mylist
mylist.append(4)  # this is going to add the element 4 at the end of the list
mylist
````

we use `.extend()` to add items of another list to other list

````python
mylist = [1, 2, 3]
list2 = [4, 6, 7]
#if we want to add elements of list2 to mylist, to make mylist = [1, 2, 3, 4, 6,7]
mylist.extend(list2)
print(mylist)
````

````python
#be carefull in using append() for adding list into list
mylist = [1, 2, 3]
list2 = [4, 6, 7]
mylist.append(list2)
print(mylist) #as you can see, the list is entirelly added to the list, not individual elements
````

we use `.insert()` to add an element at specific index not at the end

````python
mylist = [1, 2, 3]
mylist.insert(2, 5) #here we are saying, add number 5 at index 2
print(mylist) #mylist will look like [1, 2, 5, 3] ,because index 2 becomes 5
````

### Finding the Index of an Item
we use the `.index() function` to get the index of an item


imagine we have a very long list, if you want to find **the index of the first occurance** of an item

for example:
```python
myList = [3, 8, 32, 3, 5, 6, 7, 7, 73, 2, 12, 'hello', 'hi', True, False, 21]
```

Now to find where 'hello' is we use
```python
myList.index('hello')
```

````python
myList = [3, 8, 32, 3, 5, 6, 7, 7, 73, 2, 12, 'hello', 'hi', True, False, 21]
myList.index('hello')
````

Remeber `.index()` is going to return the index of the ***`FIRST OCCURENCE of the item if the item has duplicate`***

In the above example, we know that we have the item 3 repeated 2 times(at index 0 and 3), so when we use `.index() `it will give us the index of first occurence of 3(which is 0)

````python
myList.index(3)
````

### Slicing
* Slicing is *`breaking list into subpart by using indices`*
* To do slicing we use `[starting_index: end_index]` , remember **`the value at end_index is not included in the slice`**

`Example`

```python
l = [10, 43, 2, 7]
l[0:2]
```

here we are slicing the values from index 0 to index 1 so we get `[10, 43]`

````python
l = [10, 43, 2, 7]
l[0:2]
````

````python
x = [1, 'hello', 0, 'world']
x[1:3] #slicing from index 1 to 2
````

if want everything ***from certain index up until the end***, we just use `[start_index:]`, we don't need to write the end_index

````python
x = [1, 'hello', 0, 'world']
x[2:] # this means we want everything starting from index 2
````

if want ***from the beginning until certain index***, we use `[:end_index]`, we don't need to write the beginning

````python
x[:2] #this means we want everything from the beginning till index 1
````

### Slicing with Skipping

s[3:7:2] # this means from index 3 to 6 by skipping 2 steps
```

# 6. String Formating

## String Formatting with f-strings

```python
name = "John"
age = 23

print(f"Hello, {name}!")
print(f"{name} is {age} years old.")


````python
first_name = "Sara"
last_name = "Abraham"
age = 16

# Combine them into a sentence
print(f"{first_name} {last_name} is {age} years old.")
````

````python
# This prints out "John is 23 years old."
name = "John"
age = 23
print("%s is %d years old." % (name, age))
````

````python
# Format a float number to 1 or 2 decimal places
price = 4.5678
print(f"Price (2 decimals): {price:.2f}")
print(f"Price (1 decimal): {price:.1f}")
````
