## 8. Python Data Types



Everything in Python has a **type**.

Here are some of the most common types:

| Type     | Example                        | What It Represents                      |
|----------|--------------------------------|------------------------------------------|
| `int`    | `5`, `100`, `-3`               | Whole numbers                            |
| `float`  | `3.14`, `2.0`, `-1.5`          | Numbers with decimals                    |
| `str`    | `"hello"`, `'7'`, `"42.0"`     | Text (strings)                           |
| `bool`   | `True`, `False`                | Yes/No or On/Off                         |
| `list`   | `[1, 2, 3]`, `["a", "b", "c"]` | A collection of items (like a box)       |
| `None`   | `None`                         | "Nothing" or "empty" (but still valid)   |


Remember there are more data types like tuples, dictionaries...!


````python
# Examples of different data types

a = 10                 # int
b = 3.14               # float
c = "hello"            # str
d = True               # bool
e = [1, 2, 3]          # list
f = None               # NoneType

print("a:", a, "->", type(a))
print("b:", b, "->", type(b))
print("c:", c, "->", type(c))
print("d:", d, "->", type(d))
print("e:", e, "->", type(e))
print("f:", f, "->", type(f))
````

## 13. Operations(boolean types)

Python can compare values and give answers like `True` or `False`.

### Comparison Operators:

| Operator | Symbol | Example       | Result      |
|----------|--------|---------------|-------------|
| Equals   | `==`   | `5 == 5`      | `True`      |
| Not Equal| `!=`   | `5 != 3`      | `True`      |
| Greater  | `>`    | `7 > 3`       | `True`      |
| Less     | `<`    | `4 < 2`       | `False`     |
| Greater or Equal | `>=` | `5 >= 5` | `True` |
| Less or Equal | `<=` | `2 <= 3`    | `True` |

---

### Logical Operators:

| Operator | Keyword | Example                   | Result      |
|----------|---------|---------------------------|-------------|
| AND      | `and`   | `5 > 3 and 2 < 4`          | `True`      |
| OR       | `or`    | `5 < 3 or 2 < 4`           | `True`      |
| NOT      | `not`   | `not (5 > 3)`              | `False`     |

- `and`: Both sides must be `True`
- `or`: At least one side must be `True`
- `not`: Flips the value (`True` → `False`, `False` → `True`)


````python
# Try Boolean comparisons

a = 5
b = 3

print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
````

````python
# Logical operators in action

a = 5
b = 3

print("a > 2 and b > 2:", a > 2 and b > 2)
print("a > 2 and b > 5:", a > 2 and b > 5)
print("a < 2 or b > 2:", a < 2 or b > 2)
print("not a > 2:", not a > 2)
````

### Challenge: Who Gets a Discount?

Imagine you're building a program for a movie theater.

They want to give **a discount** if:
- The person is **under 18 years old**, OR
- The person **is a student**

Your task:
1. Ask the user for their age
2. Ask if they are a student (answer: "yes" or "no")
3. Decide if they get a discount using **Boolean logic**
4. Print `True` if they get the discount, `False` if not

Use `or`, `==`, and `int()` for this challenge.

Can you figure it out? Try it in the code cell below!


````python
# Write your solution here!

# Step 1: Get age
# Step 2: Get student status
# Step 3: Check if age < 18 OR student is yes
# Step 4: Print True or False
````

````python
age = int(input("Enter your age: "))
student = input("Are you a student? (yes/no): ")

# Convert student answer to boolean
is_student = student.lower() == "yes"

# Check discount condition
can_get_discount = age < 18 or is_student

print("Eligible for discount:", can_get_discount)
````
