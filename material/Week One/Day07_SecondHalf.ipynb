## Artificial Intelligence, Machine Learning & Deep Learning

* **AI**: Any computer system that can perform tasks that normally require human intelligence. It can be rule-based or search-based. But it **never learns the rules by itself**.
  * Eg: Google Translate, Chess-playing computer.
* **Machine Learning**: A subset of AI where the computer learns patterns from data instead of being explicitly programmed.
  * Eg: Email spam filter learns from spam & non-spam examples.

* **Deep Learning**: A subset of ML that uses artificial neural networks with many layers to learn patterns from large, complex datasets.
  * Eg: Face recognition on Facebook, self-driving cars.

## Types of AI

* **Vision AI** – seeing and understanding images. Eg: Image classification, face recognition.

* **Language AI** – reading, writing, understanding speech. Eg: Chatbots, translation.

* **Decision AI** – Choosing actions based on data. Recommender systems (Netflix).

* **Creative AI** – generating art, music, stories. Eg. Image & music generation.

## How AI Learns

**Data** = `Fuel for AI.`

**Features & labels** (e.g., image of a cat + label “cat”).

**Training → Prediction → Feedback loop** (learning from mistakes).

### How Machine Learning works
* **Collect data**: images, text, numbers, etc.

* **Extract features**: **`pick out the important information`** from that data.

* **Train model**: use those features **`to learn patterns`**.

* **Predict**: use patterns to **`guess new answers`**.

## Machine Learning Categories

### 1. Supervised Learning

Learns from **labeled examples** (`input + correct answer`).

Example: `Photos of cats and dogs, each labeled “cat” or “dog.”`

**Goal**: ***`Predict the label for new, unseen data.`***

**Common algorithms:** Classification, Regression, Decision Trees, Neural Networks.

#### A. Classification
* Sorting data into **categories**.

Example:

* Classifying fruit as an apple or a banana?

* Spam detection, medical diagnosis, handwriting recognition.

#### B. Decision Tree
**Each “node” splits data by asking a question.**

### 2. Unsupervised Learning

Learns from **data without labels**: the `AI must find patterns or groups.`

Example: `Grouping customer purchase data into clusters without knowing the categories beforehand.`

**Common algorithms**: Clustering (K-means), Dimensionality Reduction.

image.png

### 3. Reinforcement Learning

Learns by **trial and error** to `maximize rewards.`

Example: `Training a robot to walk or an AI to play a video game.`

## AI Strengths & Limitations

* **Good**: Fast, handles huge data, can work 24/7, accurate (sometimes).

* **Weakness**: Bias, can make mistakes. It can’t think like a human, only as good as its data.

*`Example of bias:`* model trained **only on white cats**.
