## One more Lesson
How to make string all upper case and lowercase?


*   we use .upper() to make string uppercase
*   we use .lower() to make string lowercase



````python
name = "Henok"
#change it to upper case
print(name.upper())
#change it to lower case
print(name.lower())
````
